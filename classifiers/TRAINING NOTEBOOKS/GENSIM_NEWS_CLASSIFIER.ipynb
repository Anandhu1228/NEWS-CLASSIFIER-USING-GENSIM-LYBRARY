```python
import gensim.downloader as api

GENMOD = api.load("word2vec-google-news-300")
```

```python
import pandas as pd

df = pd.read_csv("fake_and_real_news.csv")
df
```

```python
df['label'].value_counts()
```

```python
df['label'] = df['label'].map({
    "Fake" : 0,
    "Real" : 1
})
df
```

HERE EVEN IF WE PREPROCESS(THAT IS REMOVE STOPWORD AND PUNCTUATION),IT WILL AGAIN RETURN A VECTOR OF SIZE 300. IT IS BECAUSE WE ARE GIVING A TEXT INSTEAD OF WORD.

```python
import spacy
nlp = spacy.load("en_core_web_lg")

def preprocess_and_vectorize(text):
    doc = nlp(text)
    FLTTK = []
    for i in doc:
        if i.is_stop or i.is_punct:
            continue
        FLTTK.append(i.lemma_)
    return GENMOD.get_mean_vector(FLTTK)

# GET MEAN VECTOR TAKES A LIST OF WORDS, THEN RETURN A MEAN VECTOR OF THAT 
```

```python
df['Vector'] = df['Text'].apply(lambda x: preprocess_and_vectorize(x))
df
```

```python
from sklearn.model_selection import train_test_split

x = df['Vector']
y = df['label']

x_train, x_test, y_train, y_test = train_test_split(x.values, y,test_size=0.2, random_state=10)
```

```python
x_train
```

```python
print(x_train.ndim)
print(x_train.shape)
```

```python
import numpy as np

x_train = np.stack(x_train)
x_train

# HERE WE ARE CONVERTING TO A PROPER 2D NUMPY ARRAY
```

```python
print(x_train.ndim)
print(x_train.shape)

x_test = np.stack(x_test)
```

```python
from  sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

knnm = KNeighborsClassifier(n_neighbors = 5, metric = 'euclidean')

knnm.fit(x_train, y_train)

y_pred = knnm.predict(x_test)

print(classification_report(y_test, y_pred))
print(knnm.score(x_test,y_test))
```

```python
import pickle

with open("New_True_Fake.pkl","wb") as f:
    pickle.dump(knnm,f)
```
