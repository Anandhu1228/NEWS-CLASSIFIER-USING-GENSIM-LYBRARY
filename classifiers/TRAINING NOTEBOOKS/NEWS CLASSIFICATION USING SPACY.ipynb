```python
import pandas as pd

df = pd.read_csv("Fake_Real_Data.csv")
df
```

```python
df['label'].value_counts()
```

```python
df['label'] = df['label'].map({
    "Fake" : 0,
    "Real" : 1
})
df
```

```python
import spacy
nlp = spacy.load("en_core_web_lg")
```

```python
df['Vector'] = df['Text'].apply(lambda x: nlp(x).vector)
df
```

```python
from sklearn.model_selection import train_test_split

x = df['Vector']
y = df['label']

x_train, x_test, y_train, y_test = train_test_split(x.values, y,test_size=0.2, random_state=10)
```

```python
x_train
```

```python
print(x_train.ndim)
print(x_train.shape)
```

```python
import numpy as np

x_train = np.stack(x_train)
x_train

# HERE WE ARE CONVERTING TO A PROPER 2D NUMPY ARRAY
```

```python
print(x_train.ndim)
print(x_train.shape)

x_test = np.stack(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
# USED SCALER BECAUSE NEGATIVE VALUES ARE NOT ACCEPTED INTO NAIVE BAYES
from sklearn.metrics import classification_report


scaler = MinMaxScaler()
scaled_train = scaler.fit_transform(x_train)
scaled_test = scaler.transform(x_test)

nvm = MultinomialNB()
nvm.fit(scaled_train, y_train)

y_pred = nvm.predict(scaled_test)

print(classification_report(y_test, y_pred))
print(nvm.score(scaled_test,y_test))
```

```python
from  sklearn.neighbors import KNeighborsClassifier

knnm = KNeighborsClassifier(n_neighbors = 5, metric = 'euclidean')

knnm.fit(x_train, y_train)

y_pred = knnm.predict(x_test)

print(classification_report(y_test, y_pred))
print(knnm.score(x_test,y_test))
```
